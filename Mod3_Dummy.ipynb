```python
## if you want to get / print the top ten words and their scores:
def find_top_words(values, words):
    # Create pairs of (absolute value, original value, word)
    pairs = [(abs(val), val, word) for val, word in zip(values, words)]
    
    # Sort pairs by absolute value in descending order
    sorted_pairs = sorted(pairs, key=lambda x: x[0], reverse=True)
    
    # Get the top 10 (or fewer if the list is shorter)
    top_n = min(10, len(sorted_pairs))
    result = [(pair[1], pair[2]) for pair in sorted_pairs[:top_n]]
    
    return result

# Example usage
if __name__ == "__main__":
    # Sample data
    values = scores
    words = pairs
    
    top_words = find_top_words(values, words)


    words_list = []
    print("Top words by absolute value magnitude:")
    for value, pair_words in top_words:
        if value > 0:
            word = pair_words[1]
        else:
            word = pair_words[0]
        words_list.append(word)
        
        print(f"{word}: {value}")
```
